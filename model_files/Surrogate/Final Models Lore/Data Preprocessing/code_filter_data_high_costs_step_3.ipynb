# Code to filter input data with high costs

Filter out data points with exceptionally high costs to obtain a more evenly distributed data file. 

```python
import pandas as pd

# Load the data from the Excel file
file_path = '/kaggle/input/montecarloversion1701/surrogate_data_wind_month_montecarlo_version_17-01.xlsx'
data = pd.read_excel(file_path)

print(data.shape)

# Calculate the mean and standard deviation of the "Costs" column
mean_costs = data['Costs'].mean()
std_dev_costs = data['Costs'].std()

# Define the threshold for outliers (e.g., standard deviations above the mean)
threshold = mean_costs + 0.5 * std_dev_costs

# Filter the data to remove rows where "Costs" exceed the threshold
filtered_data = data[data['Costs'] <= threshold]

print(filtered_data.shape)
print(filtered_data.head())



# Save the filtered data to a new Excel file
filtered_data.to_excel('filtered_data.xlsx', index=False)

print("Filtered data saved to 'filtered_data.xlsx'")
```
